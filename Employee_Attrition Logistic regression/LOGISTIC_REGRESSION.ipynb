# Importing  Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score
```

# load Dataset

```python
df = pd.read_csv('Employee_Attrition.csv')
print(df.head())
```

# Information of dataset

```python
df.info()
```

```python
print(df.describe())
```

# Data preprocessing

```python
df.isnull().sum()
```

# Encoding Categorical Variables

```python
df['Attrition'] = df['Attrition'].map({'Yes':1 , 'No':0})
df['Gender'] = df['Gender'].map({'Male':1 , 'Female':0})
df = pd.get_dummies(df,columns=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus','OverTime'],drop_first=True)
```

# Feature selection 

```python
features = ['Age' , 'JobSatisfaction','MonthlyIncome','YearsAtCompany','Gender','TotalWorkingYears','NumCompaniesWorked','OverTime_Yes']
X = df[features]
y = df['Attrition']
```

# Feature Scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Train-Test Split

```python
X_train , X_test , y_train , y_test = train_test_split(X_scaled , y ,test_size=0.3,random_state=42)
```

# Build and Train the logistic Regression model

```python
log_model = LogisticRegression()
log_model.fit(X_train , y_train)
```

# Model Evaluation

```python
y_pred = log_model.predict(X_test)
```

# Confusion Matrix

```python
conf_matrix = confusion_matrix(y_test , y_pred)
sns.heatmap(conf_matrix,annot=True,fmt='d',cmap="Blues")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

# Classification Report

```python
print(classification_report(y_test , y_pred))
print(f"Accuracy: {accuracy_score(y_test , y_pred) * 100:.2f}%")
```

# Feature Importance

```python
feature_importance = pd.DataFrame({'Feature':features , 'Coefficient':log_model.coef_[0]})
feature_importance.sort_values(by='Coefficient',ascending=False,inplace=True)
print(feature_importance)
```

# Unseen Data

```python
unseen_data = pd.DataFrame({
    'Age':[35],
    'JobSatisfaction':[4],
    'MonthlyIncome':[6500],
    'YearsAtCompany':[5],
    'Gender':[1],
    'TotalWorkingYears':[10],
    'NumCompaniesWorked':[2],
    'OverTime_Yes':[1]
})

unseen_data_scaled = scaler.transform(unseen_data)

unseen_prediction = log_model.predict(unseen_data_scaled)
unseen_probability = log_model.predict_proba(unseen_data_scaled)

print(f"Prediction for unseen data poit : {'Attrition' if unseen_prediction[0] == 1 else 'No Attrition'}")
print(f"Prediction Probabilities : {unseen_probability[0]} ")
```
